```python
import pickle 
import matplotlib.pyplot as plt
```

```python
f = open('F_Results.pkl','rb')
arr = pickle.load(f)
f.close()
```

```python
epoch_list = arr[0]
val_accuracy = arr[1]
val_loss = arr[2]
train_accuracy = arr[4][::10]
train_loss = arr[5][::10]
```

```python
plt.rcParams["font.family"] = "Times New Roman"
```

```python
plt.xlabel("Epochs")
plt.ylabel("Accuracy")


plt.plot(epoch_list,val_accuracy,label="Validation")
plt.plot(epoch_list,train_accuracy,label="Training")
plt.legend()
plt.title("Train and Validation accuracy over time")
```

```python
plt.xlabel("Epochs")
plt.ylabel("Loss")


plt.plot(epoch_list,val_loss,label="Validation")
plt.plot(epoch_list,train_loss,label="Training")
plt.legend()
plt.title("Train and Validation Loss over time")
```
